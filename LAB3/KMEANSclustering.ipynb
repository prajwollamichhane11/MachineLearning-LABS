```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets.samples_generator import make_blobs
```

```python
X, _ = make_blobs(n_samples=1000, centers=5, n_features=2, cluster_std=1)
```

```python
X
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=3,random_state=0).fit(X)
```

```python
# Centroids of the cluster
centr = kmeans.cluster_centers_
centr
```

```python
labels = kmeans.labels_
```

```python
# Plotting the cluster centers and the data points
plt.scatter(X[:,0],X[:,1])
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c='red', marker='x')
plt.title('Data points and cluster centroids')
plt.show()
```

```python
# Seems like the no. of clusters should be increased so should the cluster centers be
# updating on the k means cluster parameter
kmeans = KMeans(n_clusters=4,random_state=0).fit(X)
```

```python
# Plotting the cluster centers and the data points
plt.scatter(X[:,0],X[:,1])
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c='red', marker='x')
plt.title('Data points and cluster centroids')
plt.show()
```
