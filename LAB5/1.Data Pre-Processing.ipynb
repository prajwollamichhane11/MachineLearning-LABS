```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
```

```python
clickbait = pd.read_fwf("clickbait_data", header=None, names=["Clickbaits"])
nonclickbait = pd.read_fwf("non_clickbait_data", header=None, names=["Clickbaits",1,2,3,4])
```

```python
nonclickbait.drop([1,2,3,4],axis=1,inplace=True)
nonclickbait.dropna(inplace=True)
```

```python
nonclickbait.head()

nonclickbait["bait"] = np.zeros((len(nonclickbait),1),dtype=int)
```

```python
nonclickbait
```

```python
clickbait.dropna(inplace=True)
clickbait.head()
```

```python
clickbait.index
clickbait["bait"] = np.ones((len(clickbait),1),dtype=int)
```

```python
clickbait
```

```python
all_dfs = [clickbait,nonclickbait]

for df in all_dfs:
    df.columns = ["Clickbaits","bait"]
data = pd.concat(all_dfs).reset_index(drop=True)
```

```python
data.dropna(inplace=True)
```

```python
data
```

```python
data = data.sample(frac=1).reset_index(drop=True)
```

```python
data
```

```python
classification = ['Clickbaits', 'Non-Clickbaits']
count = [len(clickbait),len(nonclickbait)]
plt.figure(figsize=(5,5))
plt.pie(count, labels = classification,autopct='%1.2f%%')
plt.savefig("initialCountPIE.png")
plt.show()
```

```python
classification = ['Clickbaits', 'Non-Clickbaits']
count = [len(clickbait),len(nonclickbait)]
barPlot = plt.bar(classification,count)
barPlot[0].set_color("r")
barPlot[1].set_color("g")
plt.savefig("initialCountBAR.png")
plt.show()
```
